```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def showImages(img1, img2, titulo1, titulo2):
  figure = plt.figure(figsize= (50,50))
  row = 1
  columns = 2
  figure.add_subplot(row, columns, 1)
  plt.imshow(img1)
  plt.axis('off')
  plt.title(titulo1)
  figure.add_subplot(row, columns, 2)
  plt.imshow(img2)
  plt.axis('off')
  plt.title(titulo2)
```

* Função da 'skeletonização'
1. Ler img
2. Criar Threshold (Imagem cinza = imagem binária)
3. Erodir
4. Dilatar
5. Subtrair as duas op
6. repetir até o final do processo 

```python
# MORPH_RECT
# MORPH_CROSS



def skeleton(caminho_img):
  count = 0
  img = cv2.imread(caminho_img, 0)
  img_original_copia = img.copy()
  skeleton = np.zeros(img.shape, np.uint8) #inicialmente img 'vazia'
  ret,img = cv2.threshold(img, 127,255,0)
  elemento = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)) #mudar tamanho?
  conclusao = False

  while ( conclusao == False):
    img_erosao = cv2.erode(img, elemento)
    img_dilat = cv2.dilate(img_erosao, elemento)
    img_dilat = cv2.subtract(img, img_dilat)
    skeleton = cv2.bitwise_or(skeleton,img_dilat)
    img = img_erosao.copy()
    
    if (cv2.countNonZero(img) == 0):
      conclusao = True

    count+=1
    if (count%50 == 0):  #quantidade passos depende da qualidade da imgaem
      showImages(img_original_copia, skeleton, "Orig", "Skel")
  
  showImages(img_original_copia, skeleton, "Orig", "Skel")
```
